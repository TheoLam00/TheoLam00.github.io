```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/TheoLam00/TheoLam00.github.io/refs/heads/main/project/unem/Fred-Unemployment.csv')
```

```python
df = df[(df['state'] == 'Pennsylvania')| (df['state'] == 'New Jersey')]
#filter date between 1990-1995
df = df[(df['date'] >= '1990-01-01') & (df['date'] <= '1993-12-01')]
```

```python
import seaborn as sns
```

```python
#plot value of to state in time series
sns.lineplot(x='date', y='value', hue='state', data=df)
```
