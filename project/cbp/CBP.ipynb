```python
import pandas as pd
import zipfile as zp
import requests
from io import BytesIO
```

```python
#reading data from https://www.census.gov/data/datasets/2020/econ/cbp/2020-cbp.html into pandas
data = requests.get('https://www2.census.gov/programs-surveys/cbp/datasets/2020/cbp20st.zip')
file = zp.ZipFile(BytesIO(data.content))
df = pd.read_csv(file.open('cbp20st.txt'))
```

```python
df
```

```python
#naics of sectors
naics_name = {'11':'Agriculture, Forestry, Fishing and Hunting','21':'Mining, Quarrying, and Oil and Gas Extraction','22':'Utilities','23':'Construction','31':'Manufacturing','42':'Wholesale Trade','44':'Retail Trade','48':'Transportation and Warehousing','51':'Information','52':'Finance and Insurance','53':'Real Estate and Rental and Leasing','54':'Professional, Scientific, and Technical Services','55':'Management of Companies and Enterprises','56':'Administrative and Support and Waste Management and Remediation Services','61':'Educational Services','62':'Health Care and Social Assistance','71':'Arts, Entertainment, and Recreation','72':'Accommodation and Food Services','81':'Other Services (except Public Administration)','99':'Industries not classified'}
naics = ['11----','21----','22----','23----','31----','42----','44----','48----','51----','52----','53----','54----','55----','56----','61----','62----','71----','72----','81----','99----']
```

```python
#keep only all estisblishment of legal form of organization and total of the sector
df = df[df['lfo']=='-']
df = df[df['naics'].isin(naics)]
df['naics'] = df['naics'].str.slice(0,2)
```

```python
df = df[['fipstate','naics','emp','ap']]
df = df.rename(columns={'fipstate':'state','ap':'annual_payroll','emp':'num_employees'})
df
```
