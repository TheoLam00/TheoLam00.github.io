```python
import pandas as pd
import zipfile as zp
import requests
from io import BytesIO
```

```python
data = requests.get('https://www2.census.gov/programs-surveys/cbp/datasets/2022/cbp22st.zip')
file = zp.ZipFile(BytesIO(data.content))
df = pd.read_csv(file.open('cbp22st.txt'))
```
