```python
import pandas as pd
```

```python
spending = pd.read_csv('https://raw.githubusercontent.com/TheoLam00/TheoLam00.github.io/refs/heads/main/project/spending/spending.csv')
eCollege = pd.read_csv('https://raw.githubusercontent.com/TheoLam00/TheoLam00.github.io/refs/heads/main/project/pVote/USelectionCollege.csv')
vote = pd.read_csv('https://raw.githubusercontent.com/TheoLam00/TheoLam00.github.io/refs/heads/main/project/swingState/pVoteDemRatio-long.csv')
spending = spending.drop(columns={'Unnamed: 0'})
vote = vote.drop(columns={'Unnamed: 0'})
eCollege = eCollege.melt(id_vars=['State'], var_name='year', value_name='vote')
eCollege = eCollege.rename(columns={'State': 'state'})
```

```python
party = {'McCain, John S':'Republican','Obama, Barack':'Democrats','Romney, Mitt':'Republican','Clinton, Hillary Rodham':'Democrats','Trump, Donald J.':'Republican','Biden, Joseph R Jr':'Democrats'}
spending['party'] = spending['candidate'].map(party)
spending = spending.dropna()
spending = spending.drop(columns={'candidate'})
```

```python
eCollege['year'] = eCollege['year'].astype(int)
eCollege['vote'] = eCollege['vote'].astype(int)
spending['year'] = spending['year'].astype(int)
df = pd.merge(spending, eCollege, on=['state', 'year'])
df.dropna()
df = pd.merge(df, vote, on=['state', 'year'])
df.dropna()
df
```
